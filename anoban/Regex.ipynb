```python
import re
```

```python
with open(r"Anna Karenina - Leo Tolstoy.txt", "r", encoding = "utf-8") as handle:
    anna_karenina = handle.read()
```

```python
anna = re.compile(r"A[an]{3}[\s]{1}K[areni]+")
```

```python
re.findall(anna, anna_karenina)
```

```python
set(re.findall("lo[a-z]*?e", anna_karenina))
```

```python
# get all the Wh squestions

re.findall(r"Wh[\w\s]+[\?]", anna_karenina)
```

```python
# street names

re.findall(r"[\w]+\s[\w]*street", anna_karenina)
```

```python
# possessives

re.findall(r"[\w]+[\']s", anna_karenina)
```

```python
# Names

re.findall(r"[A-Z][a-z]+\s[A-Z][a-z]+", anna_karenina)[::100]
```

```python
# Names followed by a !

re.findall(r"[A-Z][a-z]+\!", anna_karenina)
```

```python
# times

set(re.findall(r"[\w]+ o’clock", anna_karenina))
```

```python
# fetch words flanked by underscores on either side

re.findall(r"_[\w]+_", anna_karenina)
```
